```python
dat1 = pd.DataFrame(
    {"name" : ["Park", "Moon", "Hong", "Choi"],
     "gender" : ["M", "F", "M", "F"]})
```

```python
dat2 = pd.DataFrame(
    {"name" : ["Park", "Moon", "Hong", "Kim"],
     "weight" : [70, 55, 65, 80]})
```

```python
pd.merge(dat1, dat2, on="name", how="inner") # inner join 공통된거끼리 merge
```

```python
pd.merge(dat1, dat2, on="name", how="left")
```

```python
pd.merge(dat1, dat2, on="name", how="right")
```

```python
dat1.merge(dat2, on="name", how="outer")
```

```python
pd.concat([dat1, dat2], ignore_index=True)
```

```python
pd.concat([dat1, dat2], axis = 1)
```
